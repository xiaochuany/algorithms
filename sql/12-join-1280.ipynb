## Write a solution to find the number of times each student attended each exam.

Return the result table ordered by student_id and subject_name.

```sqlite
select student_id, subject_name, count(*) as attended_exams
from examinations
group by student_id, subject_name
```

```sqlite
select * from students s cross join subjects sub
```

```sqlite
select * 
from students s cross join subjects sub
left join (
    select student_id, subject_name, count(*) as attended_exams
    from examinations
    group by student_id, subject_name
) grouped
on s.student_id=grouped.student_id and sub.subject_name=grouped.subject_name
```

```sqlite
select s.student_id, s.student_name, sub.subject_name, grouped.attended_exams
from students s cross join subjects sub
left join (
    select student_id, subject_name, count(*) as attended_exams
    from examinations
    group by student_id, subject_name
) grouped
on s.student_id=grouped.student_id and  sub.subject_name=grouped.subject_name
```

```sqlite
select s.student_id, s.student_name, sub.subject_name, ifnull(grouped.attended_exams,0) as attended_exams
from students s cross join subjects sub
left join (
    select student_id, subject_name, count(*) as attended_exams
    from examinations
    group by student_id, subject_name
) grouped
on s.student_id=grouped.student_id and  sub.subject_name=grouped.subject_name
order by s.student_id, sub.subject_name
```

## keywords

- `cross join`
- `left join` (subquery)
- `ifnull( , defaul_value)`, also works with (`case when ... then ... else ... end`). the latter is roughly equiv to `np.where` 

```sqlite
select s.student_id, s.student_name, sub.subject_name, (case when grouped.attended_exams is null then 0 else grouped.attended_exams end) as attended_exams
from students s cross join subjects sub
left join (
    select student_id, subject_name, count(*) as attended_exams
    from examinations
    group by student_id, subject_name
) grouped
on s.student_id=grouped.student_id and  sub.subject_name=grouped.subject_name
order by s.student_id, sub.subject_name
```
