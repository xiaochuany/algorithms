```python
from typing import *

def minDistance(height: int, width: int, tree: List[int], squirrel: List[int], nuts: List[List[int]]) -> int:
    minimum=float('inf')
    idx = None # store idx of the nut cloest to squirrel (choose the one cloest to tree if tie)
    res = 0
    # distance to the closest nut
    for i, (nx,ny) in enumerate(nuts):
        res += (abs(nx-tree[0])+abs(ny-tree[1]))*2
        metric = abs(squirrel[0]-nx) + abs(squirrel[1]-ny) 
        if metric<minimum: minimum=metric # look for min
    # amongst the closest nuts, find the one that's closest to the tree
    min_to_tree = float('inf')
    for i, (nx,ny) in enumerate(nuts):
        if abs(squirrel[0]-nx)+abs(squirrel[1]-ny) == minimum:
            if abs(tree[0]-nx)+abs(tree[1]-ny)<min_to_tree: min_to_tree=abs(tree[0]-nx)+abs(tree[1]-ny); idx=i
    # summary
    res = res - abs(nuts[idx][0]-tree[0]) - abs(nuts[idx][1]-tree[1]) + minimum
    return res
```

```python
def test():
    assert minDistance(5, 7, [2, 2], [4, 4], [[3, 0], [2, 5]]) == 12
    assert minDistance(5, 5, [3, 2], [0, 1],
                       [[2, 0], [4, 1], [0, 4], [1, 3], [1, 0],
                        [3, 4], [3, 0], [2, 3], [0, 2], [0, 0],
                        [2, 2], [4, 2], [3, 3], [4, 4], [4, 0],
                        [4, 3], [3, 1], [2, 1], [1, 4], [2, 4]]) == 100
    assert minDistance(7, 6, [5, 4], [3, 4],
                       [[1, 4], [3, 0], [5, 0], [6, 1], [1, 3],
                        [1, 1], [0, 3], [2, 2], [6, 3], [4, 1],
                        [2, 4], [2, 1], [1, 0], [3, 2], [4, 2],
                        [2, 5], [0, 2], [1, 2], [3, 3], [6, 2],
                        [2, 0], [5, 1], [4, 5], [5, 2], [0, 1],
                        [4, 0], [5, 3], [3, 5], [0, 5], [6, 4],
                        [5, 5], [2, 3], [3, 1], [4, 4], [4, 3]]) == 288

test()
```
